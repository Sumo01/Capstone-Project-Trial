# Cleaning Dataset

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import json
```

```python
with open('datasets/Dataset-3/intents.json') as file:
    data = json.load(file)
```

```python
intents_df=pd.DataFrame(data['intents'])
intents_df.head()
```

```python
intents_df.shape
```

```python
intent_counts = intents_df['tag'].value_counts()
print("Intent Counts:")
print(intent_counts)
```

```python
# Plot the distribution of intents
plt.figure(figsize=(10, 6))
intent_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Intents')
plt.xlabel('Intent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

# Explore the distribution of patterns
pattern_lengths = intents_df['patterns'].apply(len)
print("\nPattern Lengths:")
print(pattern_lengths.describe())
```

```python

# Plot the distribution of pattern lengths
plt.figure(figsize=(10, 6))
pattern_lengths.hist(bins=20, color='lightgreen')
plt.title('Distribution of Pattern Lengths')
plt.xlabel('Pattern Length')
plt.ylabel('Count')
plt.show()
```
